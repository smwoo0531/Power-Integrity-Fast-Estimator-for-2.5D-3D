# Run `Amytest1002_3d_f2f_wo_pkg` from YAML config

This notebook runs the Amytest-equivalent Python flow using an external YAML configuration so parameters can be changed outside code.

```python
from pathlib import Path
import os
import sys

cwd = Path.cwd().resolve()
base = cwd if (cwd / 'pdn').exists() else cwd.parent
sys.path.insert(0, str(base))

os.environ.setdefault('MPLCONFIGDIR', '/tmp/mpl')
os.environ.setdefault('PYTHONPYCACHEPREFIX', '/tmp/pycache')

CONFIG = base / 'configs/silicon_2p5d_default.yaml'
OUTPUT = base / 'results_silicon2p5d_default'

print('Project root:', base)
print('Config:', CONFIG)
print('Output:', OUTPUT)

import importlib
import pdn.api as api

importlib.reload(api)
print('Using API module:', api.__file__)
OUTPUT.mkdir(parents=True, exist_ok=True)
system, chip_list = api.run_amytest1002_3d_f2f_wo_pkg_from_config(
    config_path=str(CONFIG),
    output_dir=str(OUTPUT),
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# -----------------------------
# Data (from your table)
# -----------------------------
labels = [
    "C4",
    "Backside\nmetal+TSV",
    "PDN with Stacked\nvia (M1~M10)",
    "Micro bump",
    "PDN (M1~M10)",
]
values = [0.03, 5.27, 9.2, 0.1, 11.31939]

x = np.arange(len(labels))

# Grouping (to mimic the Excel multi-level category axis style)
# boundaries between groups (between bars): after C4, after Bottom Tier
group_separators = [0.5, 2.5]  # x positions between categories
group_labels = [
    ("Bottom Tier", 1.5),  # centered between bars 1 and 2
    ("Top Tier", 3.5),     # centered between bars 3 and 4
]

# -----------------------------
# Plot
# -----------------------------
fig, (ax, ax_lbl) = plt.subplots(
    2, 1,
    figsize=(11, 5.2),
    sharex=True,
    gridspec_kw={"height_ratios": [4.0, 1.35], "hspace": 0.0}
)

# --- Top axis: bars + grid ---
ax.bar(x, values, width=0.25, color="#404040")
ax.set_ylabel("IR Drop Contribution [mV]", fontsize=14)

# Match the example scale (0 to ~12)
ax.set_ylim(0, 12)

# Dense horizontal gridlines like Excel
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(MultipleLocator(1))
ax.grid(axis="y", which="major", color="#D9D9D9", linewidth=1.0)
ax.grid(axis="y", which="minor", color="#EFEFEF", linewidth=0.8)
ax.set_axisbelow(True)

# Style spines similar to the screenshot
for spine in ax.spines.values():
    spine.set_color("#BFBFBF")
ax.spines["bottom"].set_visible(False)

# Hide x tick labels on the top axis (we draw them in the bottom axis box)
ax.tick_params(axis="x", bottom=False, labelbottom=False)
ax.tick_params(axis="y", labelsize=12, colors="#4D4D4D")

# --- Bottom axis: category box + group labels ---
ax_lbl.set_ylim(0, 1)
ax_lbl.set_yticks([])
ax_lbl.set_xticks([])

# Box border (spines) like the Excel category area
for spine in ax_lbl.spines.values():
    spine.set_color("#BFBFBF")

# Vertical separators between groups
for s in group_separators:
    ax_lbl.axvline(s, color="#BFBFBF", linewidth=1.0)

# Category labels (upper part of the bottom box)
for xi, lab in zip(x, labels):
    ax_lbl.text(
        xi, 0.67, lab,
        ha="center", va="center",
        fontsize=14, color="#4D4D4D"
    )

# Group labels (bottom part of the bottom box)
for glab, gx in group_labels:
    ax_lbl.text(
        gx, 0.18, glab,
        ha="center", va="center",
        fontsize=14, color="#4D4D4D"
    )

ax_lbl.set_xlim(-0.5, len(labels) - 0.5)

plt.tight_layout()
plt.show()

# Optional: save to file
# fig.savefig("ir_drop_breakdown.png", dpi=300, bbox_inches="tight")
```
